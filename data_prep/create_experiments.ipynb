## Make Config Files

```python

import configparser
 
def create_config(
      antibody_pdb = 'data/TEST_ANTIBODY.pdb',
      antigen_pdb = 'data/TEST_ANTIGEN.pdb',
      reference_pdb = 'data/TEST_MATCHED.pdb',
      ambig_fname = "data/ambig_TEST.tbl",
      unambig_fname = "data/unambig_TEST.tbl",
      output_file = 'config.cfg'
                  ):
    config = configparser.ConfigParser()

    ## Read the configuration file
    config.read('antibody_antigen_template.cfg')

    ## Update the configuration
    config['main'] = {'run_dir': "TEST",
                      'mode': "batch",
                      'concat':  5,
                      'queue_limit': 100,
                      'molecules': [
                            antibody_pdb,
                            antigen_pdb
                            ]}
    
    config['rigidbody'] = {
       'tolerance': 5,
       'ambig_fname': ambig_fname,
       'unambig_fname': unambig_fname
    }

    config['caprieval_1'] = {'reference_fname': reference_pdb}
    config['caprieval_2'] = {'reference_fname': reference_pdb}

    config['flexref'] = {
       'tolerance': 5,
       'ambig_fname': ambig_fname,
       'unambig_fname': unambig_fname
    }

    config['caprieval_3'] = {'reference_fname': reference_pdb}
    config['caprieval_4'] = {'reference_fname': reference_pdb}
    
    config['emref'] = {
       'tolerance': 5,
       'ambig_fname': ambig_fname,
       'unambig_fname': unambig_fname
    }

    config['caprieval_5'] = {'reference_fname': reference_pdb}

 
    ## Write the configuration to a file
    with open(output_file, 'w') as configfile:
        config.write(configfile)

    ## Replace specific lines in confi file (HACKY FIX)
    with open(output_file, 'r') as configfile:
      cfgdata = configfile.read()
    cfgdata = cfgdata.replace('[main]', '## Antibody-Antigen Docking with HADDOCK3') \
                     .replace('[caprieval_1]', '[caprieval]') \
                     .replace('[caprieval_2]', '[caprieval]') \
                     .replace('[caprieval_3]', '[caprieval]') \
                     .replace('[caprieval_4]', '[caprieval]') \
                     .replace('[caprieval_5]', '[caprieval]') \
                                                                                                         
    ## Write the file out again
    with open(output_file, 'w') as configfile:
      configfile.write(cfgdata)
```

```python
create_config(output_file = 'TEST_config.cfg')
```
