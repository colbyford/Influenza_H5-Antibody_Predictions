# Check HADDOCK Experiment Status
(To be run after pulling experiments from the HPC environment.)

```python
import os
import pandas as pd
```

```python
## Load in Experiments Excel file
experiments = pd.read_excel('../../Experiments.xlsx', sheet_name='Experiments')

experiments.head()
```

```python
for index, experiment in experiments.iterrows():
    experiment_id = experiment['experiment_id']
    output_file_path = f'../../data/results/{experiment_id}/output/10_caprieval/capri_clt.tsv'

    ## Check if the file exists
    if not os.path.exists(output_file_path):
        experiments.at[index, 'complete'] = False
        continue
    else:
        experiments.at[index, 'complete'] = True
```

```python
experiments['complete'].value_counts(dropna=False).plot(kind="pie")
```

```python
## List of experiments that are complete
experiments[['experiment_id', 'complete']][experiments['complete'] == True]
```
